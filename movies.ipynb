```python
median_val=df["IMDb"].median()
# outliers = df1[(df1["IMDb"] < lb) | (df1["IMDb"] > ub)]
# outliers
```

```python
df1["Rotten Tomatoes"] = df1["Rotten Tomatoes"].str.replace("%", "", regex=False).astype(float).div(10)
```

```python
sns.boxplot(df1["Rotten Tomatoes"])
```

```python
m1 = df1["Rotten Tomatoes"].mean()
df1["Rotten Tomatoes"].fillna(m1, inplace=True)
df1["Rotten Tomatoes"] = df1["Rotten Tomatoes"].round(1)
```

```python
df1[["Directors","Genres","Language","Country"]].fillna("Unknown",inplace=True)
df1
```

```python
median_runtime = df1["Runtime"].median()
df1["Runtime"].fillna(median_runtime, inplace=True)
```

```python
def Category(x):
    if x <= 60:
        return "Series"
    elif x <= 120:
        return "Short Movie"
    else:
        return "Long Movie"

df1["Category"] = df1["Runtime"].apply(Category)
```
